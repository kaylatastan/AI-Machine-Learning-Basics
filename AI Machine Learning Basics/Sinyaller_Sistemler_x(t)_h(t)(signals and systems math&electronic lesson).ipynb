```python
import numpy as np
import matplotlib.pyplot as plt

# Adm fonksiyonu (u(t))
def u(t):
    return np.where(t >= 0, 1, 0)

# Zaman ekseni
t = np.linspace(-5, 5, 500)

# h(t) = u(t+2) - u(t)
h1 = u(t + 2)  # u(t+2)
h2 = u(t)      # u(t)
h_t = h1 - h2  # h(t)

# Grafik izimi
plt.figure(figsize=(12, 8))

# 1. Adm: u(t+2)
plt.subplot(3, 1, 1)
plt.plot(t, h1, label=r"$u(t+2)$", color="blue")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("1. Adm: $u(t+2)$")
plt.grid(True)
plt.legend()

# 2. Adm: u(t)
plt.subplot(3, 1, 2)
plt.plot(t, h2, label=r"$u(t)$", color="orange")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("2. Adm: $u(t)$")
plt.grid(True)
plt.legend()

# 3. Adm: h(t) = u(t+2) - u(t)
plt.subplot(3, 1, 3)
plt.plot(t, h_t, label=r"$h(t) = u(t+2) - u(t)$", color="green")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("3. Adm: $h(t)$")
plt.grid(True)
plt.legend()

# Genel dzenlemeler
plt.tight_layout()
plt.show()
```
