```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dataset examples
X = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])
y = np.array([3, 5, 7, 9]) 
```

```python
#Vectorisition
X_b = np.c_[np.ones((X.shape[0], 1)), X] #Bias terimi için bir sütun ekle
beta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y) #normal denklem

print("Katsayılar (beta): ", beta)
```

```python
y_pred = X_b.dot(beta)
print("Tahminler: ", y_pred)
```

```python
#Grafik
plt.scatter(X[:,0], y, color = 'blue', label = "Gerçek")
plt.plot(X[:,0], y_pred, color = 'red', label = "Tahmin")
plt.legend()
plt.title("Çoklu Doğrusal Regresyon")
plt.xlabel("X1")
plt.ylabel("y")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

y_poly_pred = model.predict(X_poly)
```

```python
plt.scatter(X[:, 0], y, color="black", label="Gerçek")
plt.plot(X[:, 0], y_poly_pred, color="pink", label="Polinom Tahmin")
plt.legend()
plt.title("Polinom Regresyon")
plt.show()
```
